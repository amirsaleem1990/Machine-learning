```r
train <- read.csv("/home/amir/github/Kaggle-compitations//House-Prices-Advanced-Regression-Techniques-kaggle-compitation/train.csv")
nums <- sapply(train, FUN = function(x) { is.numeric(x) })
n <- train[,nums]
n <- data.frame(sapply(n, FUN = function(x) { (x - mean(x, na.rm = T)) / sd(x, na.rm = T)}))
fun <- function(x) {
	    c(min(x, na.rm=T), max(x, na.rm=T), sd(x, na.rm=T), mean(x, na.rm=T))
    }

a = t(sapply(n, FUN = fun))
df = data.frame(a)
row.names(df) <- row.names(a)
colnames(df) <- c("Min", "Max", "Sd", "Mean")
df
```
